```python
import sys, math
from skimage import data, io, filters, color
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread, imshow
from skimage.color import rgb2gray
from collections import defaultdict
from random import shuffle
```

Part 1

```python
def domColors(im, k):
    mapp = {}
    for row in range(im.shape[0]):
        for col in range(im.shape[1]):
            key = tuple(im[row][col][:]) if len(im.shape) == 3 else im[row][col]
            if key not in mapp:
                mapp [key] = 1
            else :
                mapp [key] += 1
    sorted_mapp = [(key, mapp[key]) for key in sorted(mapp, key=mapp.get, reverse=True)]
    sorted_colors = []
    for i in range(k):
        sorted_colors.append(sorted_mapp[i][0])
    if len(im.shape) == 2:
        sorted_colors = [(c, c, c) for c in sorted_colors]
    return sorted_colors[:k]
```

```python
def displayColorPalette(im, dom_list):
    plt.figure()
    if len(im.shape) == 2:
        plt.imshow(im, cmap = "gray")
    else:
        plt.imshow(im)
    k = len(dom_list)
    # Most dominant colour
    top_im = np.array([[list(dom_list[0])]])
    plt.figure(figsize = (0.75,0.75), frameon=False)
    plt.imshow(top_im)
    # k most dominant colours palette
    plt.figure(figsize = ((0.75*k),0.75), frameon=False)
    # shuffle(dom_list)
    dom_cols_im = np.array([[list(dom_list[i]) for i in range(k)]])
    plt.imshow(dom_cols_im)
```

Part 2

a) Colour Image

```python
im = imread("./A1_resources/rose.jpeg")
dom_list = domColors(im,20)
displayColorPalette(im, dom_list)
```

b) Black and White Image

```python
im = imread("./A1_resources/lena.bmp")
dom_list = domColors(im,20)
displayColorPalette(im, dom_list)
```
