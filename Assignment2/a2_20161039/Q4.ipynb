```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, util
from scipy.ndimage import filters
from skimage.io import imread, imshow
from skimage.util import img_as_ubyte, img_as_float
from skimage.color import rgb2gray
```

```python
def rescale(img):
    rescaled_img = img - img.min()
    rescaled_img *= 255
    rescaled_img *= 1./(img.max() - img.min())
    return rescaled_img.astype('uint8')
```

```python
laplace = np.array([[0, -1, 0], 
                 [-1, 4, -1], 
                 [0, -1, 0]])

laplace_diag = np.array([[-1, -1, -1], 
                 [-1, 8, -1], 
                 [-1, -1, -1]])

prewitt_h = np.array([[1, 1, 1],
                    [0, 0, 0],
                    [-1, -1, -1]])

prewitt_v = np.array([[-1, 0, 1],
                    [-1, 0, 1],
                    [-1, 0, 1]])

roberts_h = np.array([[1, 0],
                    [0, -1]])

roberts_v = np.array([[0, 1],
                    [-1, 0]])

sobel_h = np.array([[1, 2, 1],
                    [0, 0, 0],
                    [-1, -2, -1]])

sobel_v = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

filter_type = [laplace, laplace_diag, prewitt_h, prewitt_v, roberts_h, roberts_v, sobel_h, sobel_v]
```

# Part 1

```python
barb = imread('./A2_resources/barbara.jpg')
barb = img_as_ubyte(barb)
imshow(barb)
for filt in filter_type:
    plt.figure()
    plt.imshow(rescale(np.absolute(filters.convolve(barb.astype(np.float), filt))), cmap="gray")
```

## Observations
### Laplacian Filter 
A second order derivative function which only captures sharp intensity changes and not gradients. Here, it captures all the edges.
### Diagonal Laplacian Filter
Just like the Laplacian filter, but it also captures edges that are not strictly vertical or horizontal better than the standard Laplace filter.
### Horizontal Prewitt Filter 
It detects horizontal edges - drastic intensity changes in the vertical direction. Features such as the lips, eyebrows etc. and other horizontal outlines are more prominent.
### Vertical Prewitt Filter 
It detects vertical edges - drastic intensity changes in the horizonatal direction. The nose is clearer and the gradient across the face is more prominent. The table leg is clearer.
### Horizontal Roberts Filter
It captures intensity changes in the top left-bottom right direction. The left leg and the nose are prominent.
### Vertical Roberts Filter 
It captures intensity changes in the top right-bottom left direction. The eyes, left hand and right leg are prominent.
### Horizontal Sobel Filter 
It produces results very similar to the Horizontal Prewitt filter. It gives more importance to pixels that are closer to the centre pixel  as compared to the Horizontal Prewitt filter.
### Vertical Sobel Filter 
It produces results very similar to the Vertical Prewitt filter. It gives more importance to pixels that are closer to the centre pixel as compared to the Vertical Prewitt filter.
