```python
from skimage.io import imread, imshow
from skimage.util import pad, img_as_ubyte, img_as_float
from skimage.color import rgb2gray
from scipy.ndimage import filters
import numpy as np
import matplotlib.pyplot as plt
import time
```

## Part 1

```python
def standard_low_pass_filter(img, k):
    fil = np.ones((k, k))
    fil = (1.0/(k * k)) * fil
    p = int((k - 1)/2)
    new_img = np.zeros((img.shape[0], img.shape[1]))
    img = pad(img, p, mode='constant')
    
    for i in range(p, img.shape[0] - p):
        for j in range(p, img.shape[1] - p):
            img_window = img[i - p:i + p + 1, j - p:j + p + 1]
            window_sum = 0
            for x in range(k):
                for y in range(k):
                    window_sum += fil[x][y] * img_window[x][y]
            new_img[i - p][j - p] = window_sum
    
    return new_img
```

```python
sky2 = img_as_ubyte(rgb2gray(imread('./A2_resources/city.jpg')))
sky2 = rgb2gray(sky2)
sky2 = img_as_ubyte(sky2)
smoothed_sky2 = standard_low_pass_filter(sky2, 3)
plt.figure(figsize=(10, 10))
plt.imshow(sky2, cmap="gray")
plt.figure(figsize=(10, 10))
plt.imshow(smoothed_sky2, cmap="gray")
```

## Part 2

```python
def efficient_low_pass_filter(img, k):
    fac = 1.0/(k * k)
    fil = fac * np.ones((k, k))
    p = int((k - 1)/2)
    new_img = np.zeros((img.shape[0], img.shape[1]))
    img = pad(img, p, mode='constant')
    
    avg = 0
    for i in range(p, img.shape[0] - p):
        if (i - p) % 2 != 0:
            for j in range(img.shape[1] - p - 1, p - 1, -1):
                if j == img.shape[1] - p - 1:
                    avg += fac * (np.sum(img[i + p, j - p:j + p + 1])) - fac * (np.sum(img[i - p - 1, j - p:j + p + 1]))
                    new_img[i - p, j - p] = avg
                else:
                    avg += fac * np.sum(img[i - p:i + p + 1, j - p]) - fac * np.sum(img[i - p:i + p + 1, j + p + 1])
                    new_img[i - p, j - p] = avg
        else:
            for j in range(p, img.shape[1] - p):
                if i == p and j == p:
                    img_window = img[i - p:i + p + 1, j - p:j + p + 1]
                    avg += np.sum(np.multiply(fil, img_window))
                    new_img[i - p, j - p] = avg
                else:
                    if j == p:
                        avg += fac * np.sum(img[i + p, j - p:j + p + 1]) - fac * np.sum(img[i - p - 1, j - p:j + p + 1])
                        new_img[i - p, j - p] = avg
                    else:
                        avg += fac * np.sum(img[i - p:i + p + 1, j + p]) - fac * np.sum(img[i - p:i + p + 1, j - p - 1])
                        new_img[i - p, j - p] = avg

    return new_img
```

```python
sky2 = imread('./A2_resources/city.jpg')
sky2 = rgb2gray(sky2)
sky2 = img_as_ubyte(sky2)
smoothed_sky2 = efficient_low_pass_filter(sky2, 5)
plt.figure(figsize=(10, 10))
plt.imshow(sky2, cmap="gray")
plt.figure(figsize=(10, 10))
plt.imshow(smoothed_sky2, cmap="gray")
```

```python
city = img_as_ubyte(rgb2gray(imread('./A2_resources/city.jpg')))
cat = img_as_ubyte(rgb2gray(imread('./A2_resources/cat.jpeg')))
house = img_as_ubyte(rgb2gray(imread('./A2_resources/house.jpg')))
x = []
y1 = []
y1_eff = []
y2 = []
y2_eff = []
y3 = []
y3_eff = []

for k in range(3, 14, 2):
    x.append(k)
    start = time.time()
    smoothed = standard_low_pass_filter(city, k)
    end = time.time()
    y1.append(end - start)
    start = time.time()
    smoothed = efficient_low_pass_filter(city, k)
    end = time.time()
    y1_eff.append(end - start)
    
    start = time.time()
    smoothed = standard_low_pass_filter(cat, k)
    end = time.time()
    y2.append(end - start)
    start = time.time()
    smoothed = efficient_low_pass_filter(cat, k)
    end = time.time()
    y2_eff.append(end - start)

    start = time.time()
    smoothed = standard_low_pass_filter(house, k)
    end = time.time()
    y3.append(end - start)
    start = time.time()
    smoothed = efficient_low_pass_filter(house, k)
    end = time.time()
    y3_eff.append(end - start)

plt.plot(x, y1, 'r', label="Standard - city")
plt.plot(x, y1_eff, 'r--', label="Efficient")
plt.plot(x, y2, 'g', label="Standard - cat")
plt.plot(x, y2_eff, 'g--', label="Efficient")
plt.plot(x, y3, 'b', label="Standard - house")
plt.plot(x, y3_eff, 'b--', label="Efficient")

plt.legend()
plt.show()
```

## Part 3

```python
def median_filter(img, k):
    p = int((k - 1)/2)
    new_img = np.zeros(img.shape)
    img = pad(img, p, mode='constant')
    
    for i in range(p, img.shape[0] - p):
        for j in range(p, img.shape[1] - p):
            img_window = img[i - p:i + p + 1, j - p:j + p + 1]
            values = []
            for x in range(img_window.shape[0]):
                for y in range(img_window.shape[1]):
                    values.append(img_window[x, y])
                    
#             for x in range(len(values)):
#                 for y in range(x + 1, len(values)):
#                     if values[y] < values[x]:
#                         temp = values[x]
#                         values[x] = values[y]
#                         values[y] = temp
            values = sorted(values)
            ind = int((k ** 2 - 1)/2)
            new_img[i - p][j - p] = values[ind]
    return new_img    
```

```python
def efficient_median_filter(img, k):
    new_img = np.zeros((img.shape[0], img.shape[1]))
    p = int((k - 1)/2)
    img = pad(img, p, mode='constant')
    histogram = [0 for _ in range(256)]
    median_ind = int((k ** 2 - 1)/2)
    
    for i in range(p, img.shape[0] - p):
        if (i - p) % 2 != 0:
            for j in range(img.shape[1] - p - 1, p - 1, -1):
                if j != img.shape[1] - p - 1:
                    new_vals = img[i - p:i + p + 1, j - p]
                    prev_vals = img[i - p:i + p + 1, j + p + 1]
                    
                    for x in range(k):
                        histogram[new_vals[x]] += 1
                    for x in range(k):
                        histogram[prev_vals[x]] -= 1                        
                    sort_win = []
                    for x in range(256):
                        sort_win.extend([x] * histogram[x])
                    new_img[i - p][j - p] = sort_win[median_ind]                                

                else:
                    new_vals = img[i + p, j - p:j + p + 1]
                    prev_vals = img[i - p - 1, j - p:j + p + 1]
                    for x in range(k):
                        histogram[new_vals[x]] += 1
                    for x in range(k):
                        histogram[prev_vals[x]] -= 1                        
                    sort_win = []
                    for x in range(256):
                        sort_win.extend([x] * histogram[x])
                    new_img[i - p][j - p] = sort_win[median_ind]                       
                                   
        else:
            for j in range(p, img.shape[1] - p):
                if i != p or j != p:
                    if j != p:
                        prev_vals = img[i - p:i + p + 1, j - p - 1]
                        new_vals = img[i - p:i + p + 1, j + p]
                        for x in range(k):
                            histogram[prev_vals[x]] -= 1
                        for x in range(k):
                            histogram[new_vals[x]] += 1
                        sort_win = []
                        for x in range(256):
                            sort_win.extend([x] * histogram[x])
                        new_img[i - p, j - p] = sort_win[median_ind]     
                    
                    else:
                        prev_vals = img[i - p - 1, j - p:j + p + 1]
                        new_vals = img[i + p, j - p:j + p + 1]
                        for x in range(k):
                            histogram[prev_vals[x]] -= 1
                        for x in range(k):
                            histogram[new_vals[x]] += 1
                        sort_win = []
                        for x in range(256):
                            sort_win.extend([x] * histogram[x])
                        new_img[i - p, j - p] = sort_win[median_ind]                         
                        
                else:
                    img_window = img[i - p:i + p + 1, j - p:j + p + 1]
                    for x in range(k):
                        for y in range(k):
                            histogram[img_window[x, y]] += 1
                    sort_win = []
                    for x in range(256):
                        sort_win.extend([x] * histogram[x])
                    new_img[i - p, j - p] = sort_win[median_ind]                                         

    return new_img
```

```python
sky2 = imread('./A2_resources/city.jpg')
sky2 = rgb2gray(sky2)
sky2 = img_as_ubyte(sky2)

smoothed_sky2 = efficient_median_filter(sky2, 7)
plt.figure(figsize=(10, 10))
plt.imshow(sky2, cmap="gray")
plt.figure(figsize=(10, 10))
plt.imshow(smoothed_sky2, cmap="gray")
```
