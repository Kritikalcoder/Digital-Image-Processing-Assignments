```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, util
from scipy.ndimage import filters
from skimage.io import imread, imshow
from skimage.util import img_as_ubyte, img_as_float
from skimage.color import rgb2gray
```

## Part 1

```python
# Inputting images
sky = imread('./A2_resources/sky.jpeg')
sky = img_as_ubyte(rgb2gray(sky))
sky2 = imread('./A2_resources/sky2.jpg')
sky2 = img_as_ubyte(rgb2gray(sky2))
```

```python
# filter A
filterA = 1/16.0 * np.array([[1, 2, 1], 
                        [2, 4, 2],
                        [1, 2, 1]])
filterA_mod = 1/9.0 * np.array([[1, 2, 1],
                        [2, 4, 2],
                        [1, 2, 1]])
```

```python
# filter B
filterB = 1/256.0 * np.array([[1, 4, 6, 4, 1],
                        [4, 16, 24, 16, 4],
                        [6, 24, 36, 24, 6],
                        [4, 16, 24, 16, 4],
                        [1, 4, 6, 4, 1]])
filterB_mod = 1/150.0 * np.array([[1, 4, 6, 4, 1],
                        [4, 16, 24, 16, 4],
                        [6, 24, 36, 24, 6],
                        [4, 16, 24, 16, 4],
                        [1, 4, 6, 4, 1]])
```

```python
# filter C
filterC = 1/8.0 * np.array([[-1, 0, 1],
                        [-2, 0, 2],
                        [-1, 0, 1]])
filterC_mod = 1/15.0 * np.array([[-2, 0, 3],
                    [-2, 0, 3],
                    [-2, 0, 3]])
```

```python
def rescale(img):
    rescaled_img = img - img.min()
    rescaled_img *= 255
    rescaled_img *= 1./(img.max() - img.min())
    return rescaled_img.astype('uint8')
```

```python
sky_A = filters.convolve(sky.astype(np.float), filterA)
sky_A = np.clip(sky_A, a_min=0, a_max=255).astype(np.uint8)

sky_A_mod = filters.convolve(sky.astype(np.float), filterA_mod)
sky_A_sc = rescale(sky_A_mod)
sky_A_mod = np.clip(sky_A_mod, a_min=0, a_max=255).astype(np.uint8)

sky_B = filters.convolve(sky.astype(np.float), filterB)
sky_B = np.clip(sky_B, a_min=0, a_max=255).astype(np.uint8)

sky_B_mod = filters.convolve(sky.astype(np.float), filterB_mod)
sky_B_sc = rescale(sky_B_mod)
sky_B_mod = np.clip(sky_B_mod, a_min=0, a_max=255).astype(np.uint8)

sky_C = filters.convolve(sky.astype(np.float), filterC)
sky_C = np.clip(sky_C, a_min=0, a_max=255).astype(np.uint8)

sky_C_mod = filters.convolve(sky.astype(np.float), filterC_mod)
sky_C_sc = rescale(sky_C_mod)
sky_C_mod = np.clip(sky_C_mod, a_min=0, a_max=255).astype(np.uint8)
```

```python
sky2_A = filters.convolve(sky2.astype(np.float), filterA)
sky2_A = np.clip(sky2_A, a_min=0, a_max=255).astype(np.uint8)

sky2_A_mod = filters.convolve(sky2.astype(np.float), filterA_mod)
sky2_A_sc = rescale(sky2_A_mod)
sky2_A_mod = np.clip(sky2_A_mod, a_min=0, a_max=255).astype(np.uint8)

sky2_B = filters.convolve(sky2.astype(np.float), filterB)
sky2_B = np.clip(sky2_B, a_min=0, a_max=255).astype(np.uint8)

sky2_B_mod = filters.convolve(sky2.astype(np.float), filterB_mod)
sky2_B_sc = rescale(sky2_B_mod)
sky2_B_mod = np.clip(sky2_B_mod, a_min=0, a_max=255).astype(np.uint8)

sky2_C = filters.convolve(sky2.astype(np.float), filterC)
sky2_C = np.clip(sky2_C, a_min=0, a_max=255).astype(np.uint8)

sky2_C_mod = filters.convolve(sky2.astype(np.float), filterC_mod)
sky2_C_sc = rescale(sky2_C_mod)
sky2_C_mod = np.clip(sky2_C_mod, a_min=0, a_max=255).astype(np.uint8)
```

```python
plt.figure()
plt.imshow(sky, cmap="gray")

sky_res = np.hstack((sky_A, sky_B, sky_C))
plt.figure(figsize=(40, 40))
plt.imshow(sky_res, cmap="gray")

sky_mod_res = np.hstack((sky_A_mod, sky_B_mod, sky_C_mod))
plt.figure(figsize=(40, 40))
plt.imshow(sky_mod_res, cmap="gray")

sky_sc_res = np.hstack((sky_A_sc, sky_B_sc, sky_C_sc))
plt.figure(figsize=(40, 40))
plt.imshow(sky_sc_res, cmap="gray")
```

```python
plt.figure()
plt.imshow(sky, cmap="gray")

sky2_res = np.hstack((sky2_A, sky2_B, sky2_C))
plt.figure(figsize=(40, 40))
plt.imshow(sky2_res, cmap="gray")

sky2_mod_res = np.hstack((sky2_A_mod, sky2_B_mod, sky2_C_mod))
plt.figure(figsize=(40, 40))
plt.imshow(sky2_mod_res, cmap="gray")

sky2_sc_res = np.hstack((sky2_A_sc, sky2_B_sc, sky2_C_sc))
plt.figure(figsize=(40, 40))
plt.imshow(sky2_sc_res, cmap="gray")
```

## Part 2

If we don't normalize filters to sum up to 1, the final intensities end up being much higher than intensities of the original image. Further, there is a possibility that the intensities end up being greater than 255, the logical upper bound of the intensity range, which results in the bright parts of the image becoming white.

When we scale the new values to the 0 to 255 range (as opposed to clipping them at 0 and 255), the bright values get a large share of the spectrum and hence the contrast is good for bright areas. But the spectrum for the dark areas is compressed and the contrast becomes poor for dark intensities.

For filters that sum up to 0 (generally sharpening filters), if the values don't sum up to 0, even areas that have uniform intensity values get non-zero values in the filtered images. 
As an example, this defeats the purpose of edge-detecting filters.

## Part 3

The filters that sum up to 1 are averaging filters where the intensity value of the centre pixel is a weighted average of the neighboring pixels' intensities. By making them sum up to 1, we make sure that no intensity value is higher than 255.

The filters that sum up to 0 are derivative filters - they detect gradients and edges. So, in an area of uniform intensity, they do not give any signal - they sum up to 0 (no information).
