```python
import sys, math, os
from skimage import data, io, filters, color, img_as_ubyte, transform, img_as_float
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread, imshow
from skimage.color import rgb2gray
from collections import defaultdict
from scipy import misc
```

```python
def BinarizeImage(im, method, window=None):
    if len(im.shape) == 2:
        im = img_as_float(im)
    else:
        im = img_as_float(color.rgb2gray(im))
    
    if method == "otsu":
        thresh_im = im > filters.threshold_otsu(im)
        return img_as_ubyte(thresh_im)
    
    elif method == "adaptive":
        thresh_im = im > filters.threshold_local(im, window)
        return img_as_ubyte(thresh_im)
```

```python
im1 = imread('./A1_resources/Q9/palm-leaf-1.jpg')

plt.figure(figsize=(30,30))
plt.imshow(im1)

plt.figure(figsize=(30, 30))
plt.imshow(BinarizeImage(im1, "otsu"), cmap="gray")

plt.figure(figsize=(30, 30))
plt.imshow(BinarizeImage(im1, "adaptive", window=205), cmap="gray")
```

```python
im2 = imread('./A1_resources/Q9/palm-leaf-2.jpg')

plt.figure(figsize=(10,10))
plt.imshow(im2)

plt.figure(figsize=(10, 10))
plt.imshow(BinarizeImage(im2, "otsu"), cmap="gray")

plt.figure(figsize=(10, 10))
plt.imshow(BinarizeImage(im2, "adaptive", window=205), cmap="gray")
```

Global Thresholding Vs Locally Adaptive Thresholding:  
Global methods perform better in high contract images, but worse in noisy images - local methods overcome this problem due to localization of pixels, avoiding noise.  
  
What are the choices that need to be made for locally adaptive thresholding ?  
Ans: The window size and the filter type.
