```python
import sys, math
from skimage import data, io, filters, color
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread, imshow
from skimage.color import rgb2gray
from collections import defaultdict
```

```python
def plotHistogram(img) :
    mapp = [0] * 256 
    for row in range(img.shape[0]):
        for col in range(img.shape[1]):
            key = img[row][col]
            mapp [key] += 1
    total = sum(mapp)
    mapp = [(val * 1.0) / total for val in mapp]
    wid = 1/1.5
    plt.figure()
    plt.bar(range(256), mapp)
    plt.show()
```

```python
low1 = imread("./A1_resources/Q7/low_1.jpg")[:400,:550,0]
low2 = imread("./A1_resources/Q7/low_2.jpg")[:400,:550,0]
low3 = imread("./A1_resources/Q7/low_3.jpg")[:400,:550,0]
high1 = imread("./A1_resources/Q7/high_1.jpg")[:400,:550,0]
high2 = imread("./A1_resources/Q7/high_2.jpg")[:400,:550,0]
high3 = imread("./A1_resources/Q7/high_3.jpg")[:400,:550,0]
```

1. High Contrast Images:

```python
plt.figure()
plt.imshow(high1, cmap = "gray")
plotHistogram(high1)
plt.figure()
plt.imshow(high2, cmap = "gray")
plotHistogram(high2)
plt.figure()
plt.imshow(high3, cmap = "gray")
plotHistogram(high3)
```

2. Low Contrast Images:

```python
plt.figure()
plt.imshow(low1, cmap = "gray")
plotHistogram(low1)
plt.figure()
plt.imshow(low2, cmap = "gray")
plotHistogram(low2)
plt.figure()
plt.imshow(low3, cmap = "gray")
plotHistogram(low3)
```

Comparison: 

Low contrast images have a relatively more uniform distribution over a subset of the range (locally close values) closer towards the bright end of the spectrum than the low end of the spectrum.

High contrast images exhibit a valley distribution - the peaks of the histogram are separated by a range of pixels - implying that it will be easier to "contrast" the dark parts of the image from the bright parts.
