```python
import sys, math, os
from skimage import data, io, filters, color, img_as_ubyte, transform
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread, imshow
from skimage.color import rgb2gray
from collections import defaultdict
from scipy import misc
from math import sqrt
```

Part 1:

```python
def ChromaKey(fg, bg, keyColor, limit):
    res = np.zeros(fg.shape, dtype=np.uint8)
    for row in range(fg.shape[0]):
        for col in range(fg.shape[1]):
            dist_r = ((fg[row][col][0] - keyColor[0]) ** 2) * 2
            dist_g = ((fg[row][col][1] - keyColor[1]) ** 2) * 4
            dist_b = ((fg[row][col][2] - keyColor[2]) ** 2) * 3
            color_dist = math.sqrt( (dist_r + dist_g + dist_b) / 9)
            if color_dist > limit:
                res[row,col,:] = fg[row,col,:]
            else :
                res[row,col,:] = bg[row,col,:]  
    temp = np.hstack((bg,fg))
    temp = np.hstack((temp,res)) 
    imshow(temp)
```

Part 2:

```python
rose = imread('./A1_resources/Q2/rose.jpeg')
# print(rose[0,0,:])
rose_bg = imread('./A1_resources/Q2/spider.jpg')[:, :, :3]
ChromaKey(rose, rose_bg, (0, 0, 0), 30)
```

Part 3:

```python
zebra = imread('./A1_resources/Q2/zebra.jpg')
# print(zebra[0,0,:])
zebra_bg = imread('./A1_resources/Q2/zebra_bg.jpg')
zebra_fg1 = img_as_ubyte(transform.rescale(zebra, 0.25))
zebra_bg1 = img_as_ubyte(transform.rescale(zebra_bg, 0.25))[:, :, :3]
ChromaKey(zebra_fg1, zebra_bg1, (52, 255, 6), 100)
```
