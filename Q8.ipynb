```python
import sys, math
from skimage import data, io, filters, color
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread, imshow
from skimage.color import rgb2gray
from collections import defaultdict
```

```python
def BitPlanes(im):
    if len(im.shape) == 3:
        im = im[:,:,0]
    res_img = np.zeros(shape = (im.shape[0], im.shape[1], 8))
    for row in range(im.shape[0]):
        for col in range(im.shape[1]):
            temp = '{0:08b}'.format(im[row][col])
            res_img[row][col] = [int(i) for i in str(temp)]
            
    for i in range(8):
        plt.figure()
        plt.imshow(res_img[:,:,i], cmap = "gray")
```

```python
im = imread("./A1_resources/Q8/lena.bmp")
imshow(im)
BitPlanes(im)
```

Observation:

The plane with the most information is the Most Significant Bit plane, i.e., the plane corresponding to the 8th bit ($2^7 = 128$). This is because that plane tells you whether the intensity value of the pixel lies between 0 to 127 or between 128 to 225.

The plane with the least amount of information is the Least Significant Bit Plane, i.e., the plane corresponding to the 1st bit ($2^0 = 1$). This is because that plane tells you whether the intensity value of the pixel is odd or even, which is not of much use, due to locality of colour wrt intensity values.
