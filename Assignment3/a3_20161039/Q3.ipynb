```python
import numpy as np
from numpy import abs, log, rot90, hstack, fft
from numpy.fft import fftshift, fft2
from skimage import io, util, color
from skimage.io import imread
from skimage.util import img_as_ubyte
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
```

Part 1

```python
im64 = imread('./images/Kristine.jpg')
im64 = img_as_ubyte(rgb2gray(im64))

im512 = np.zeros((512, 512), dtype=np.uint8)
im256 = np.zeros((256, 256), dtype=np.uint8)
im128 = np.zeros((128, 128), dtype=np.uint8)

im512[:64, :64] = im64
im256[:64, :64] = im64
im128[:64, :64] = im64

plt.figure()
plt.imshow(im512, cmap="gray")
plt.figure()
plt.imshow(im256, cmap="gray")
plt.figure()
plt.imshow(im128, cmap="gray")
plt.figure()
plt.imshow(im64, cmap="gray")
```

Part 2

```python
im512_fft = fftshift(fft2(im512))
im256_fft = fftshift(fft2(im256))
im128_fft = fftshift(fft2(im128))
im64_fft = fftshift(fft2(im64))

im512_mag = abs(im512_fft)
im256_mag = abs(im256_fft)
im128_mag = abs(im128_fft)
im64_mag = abs(im64_fft)

plt.figure()
plt.imshow(im512_mag, cmap="gray")
plt.figure()
plt.imshow(im256_mag, cmap="gray")
plt.figure()
plt.imshow(im128_mag, cmap="gray")
plt.figure()
plt.imshow(im64_mag, cmap="gray")
```

Explanation:    
Neither does the location nor does the amplitude change, but both the peak location and peak amplitude become more accurate in the zero-padded version, and this accuracy increases with the amount of padding, as the interpolation power increases. This leads to a smoothening of the frequencies. Zero padding just increases the sample density in the frequency domain, as the inverse fourier transform is scaled by a larger number (1/N vs 1/(N+M))
