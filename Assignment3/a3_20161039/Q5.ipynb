```python
import numpy as np
from numpy import abs, log, rot90, hstack, fft
from numpy.fft import fftshift, fft2, ifft2, ifftshift
from skimage import io, util, color, measure
from skimage.io import imread
from skimage.measure import block_reduce
import matplotlib.pyplot as plt
```

```python
checker = imread('./images/checkerboard.png')

checker_1 = block_reduce(checker, (1, 1))
checker_2 = block_reduce(checker, (2, 2))
checker_4 = block_reduce(checker, (4, 4))
checker_8 = block_reduce(checker, (8, 8))
checker_16 = block_reduce(checker, (16, 16))

plt.figure()
plt.imshow(checker_1, cmap="gray")
plt.figure()
plt.imshow(checker_2, cmap="gray")
plt.figure()
plt.imshow(checker_4, cmap="gray")
plt.figure()
plt.imshow(checker_8, cmap="gray")
plt.figure()
plt.imshow(checker_16, cmap="gray")
```

```python
checker_32 = block_reduce(checker, (32, 32))
checker_15 = block_reduce(checker, (15, 15))

plt.figure()
plt.imshow(checker_32, cmap="gray")
plt.figure()
plt.imshow(checker_15, cmap="gray")
```

```python
checker1_fft = fftshift(fft2(checker_1))
checker1_mag = abs(checker1_fft)
res1 = ifft2(ifftshift(checker1_fft)).astype("uint8")

checker2_fft = fftshift(fft2(checker_2))
checker2_mag = abs(checker2_fft)
res2 = ifft2(ifftshift(checker2_fft)).astype("uint8")

checker4_fft = fftshift(fft2(checker_4))
checker4_mag = abs(checker4_fft)
res4 = ifft2(ifftshift(checker4_fft)).astype("uint8")

plt.figure()
plt.imshow(log(checker1_mag + 1), cmap="gray")
plt.figure()
plt.imshow(res1, cmap="gray")

plt.figure()
plt.imshow(log(checker2_mag + 1), cmap="gray")
plt.figure()
plt.imshow(res2, cmap="gray")

plt.figure()
plt.imshow(log(checker4_mag + 1), cmap="gray")
plt.figure()
plt.imshow(res4, cmap="gray")
```

```python
checker_fft = fftshift(fft2(checker_8))
checker_mag = abs(checker_fft)
res = ifft2(ifftshift(checker_fft)).astype("uint8")

plt.figure()
plt.imshow(log(checker_mag + 1), cmap="gray")
plt.figure()
plt.imshow(res, cmap="gray")
```

Observation:  
We are able to reconstruct the image if we sample it in powers of 2, upto the value 8.  
8 is the critical sampling rate in this checkerboard, along both directions - horizontal and vertical.  
Nx = Ny = 8.

```python
checker16_fft = fftshift(fft2(checker_16))
checker16_mag = abs(checker16_fft)
res16 = ifft2(ifftshift(checker16_fft)).astype("uint8")

checker32_fft = fftshift(fft2(checker_32))
checker32_mag = abs(checker32_fft)
res32 = ifft2(ifftshift(checker32_fft)).astype("uint8")

plt.figure()
plt.imshow(log(checker16_mag + 1), cmap="gray")
plt.figure()
plt.imshow(res16, cmap="gray")

plt.figure()
plt.imshow(log(checker32_mag + 1), cmap="gray")
plt.figure()
plt.imshow(res32, cmap="gray")
```
