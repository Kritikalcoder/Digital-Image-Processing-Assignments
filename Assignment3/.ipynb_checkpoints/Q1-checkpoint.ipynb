# Q1

```python
import numpy as np
from numpy import abs, log, rot90, hstack, fft
from numpy.fft import fftshift, fft2
from skimage import io, util, color
from skimage.io import imread
import matplotlib.pyplot as plt
```

Part 1

```python
# Image 1

pattern_input = imread('./images/pattern.jpg')[:,:,0]
pattern_fft = fft2(pattern_input)
pattern_fft = fftshift(pattern_fft)
pattern_mag = abs(pattern_fft)
pattern_log_mag = log(pattern_mag + 1)

plt.figure()
plt.imshow(pattern_input, cmap="gray")
plt.figure()
plt.imshow(pattern_mag, cmap="gray")
plt.figure()
plt.imshow(pattern_log_mag, cmap="gray")

# Image 2

stripes_input = imread('./images/stripes.gif')
stripes_fft = fft2(stripes_input)
stripes_fft = fftshift(stripes_fft)
stripes_mag = abs(stripes_fft)
stripes_log_mag = log(stripes_mag + 1)

plt.figure()
plt.imshow(stripes_input, cmap="gray")
plt.figure()
plt.imshow(stripes_mag, cmap="gray")
plt.figure()
plt.imshow(stripes_log_mag, cmap="gray")
```

Part 2

```python
# Image 1

pattern_mod = log(pattern_input + 1)

plt.figure()
plt.imshow(pattern_input, cmap="gray")
plt.figure()
plt.imshow(pattern_mod.astype(np.uint8), cmap="gray")
plt.figure()
plt.imshow(pattern_mod.astype(np.uint8), cmap="gray", vmin=0, vmax=255)

# Image 2

stripes_mod = log(stripes_input + 1)

plt.figure()
plt.imshow(stripes_input, cmap="gray")
plt.figure()
plt.imshow(stripes_mod.astype(np.uint8), cmap="gray")
plt.figure()
plt.imshow(stripes_mod.astype(np.uint8), cmap="gray", vmin=0, vmax=255)
```

Part 3

```python
# Image 1

pattern_naive = abs(fft2(fft2(pattern_input)))
pattern_trans = abs(fft2(rot90(fft2(pattern_input), 2)))

plt.figure()
plt.imshow(pattern_input, cmap="gray")
plt.figure()
plt.imshow(pattern_naive, cmap="gray")
plt.figure()
plt.imshow(pattern_trans, cmap="gray")

# Image 2

stripes_naive = abs(fft2(fft2(stripes_input)))
stripes_trans = abs(fft2(rot90(fft2(stripes_input), 2)))

plt.figure()
plt.imshow(stripes_input, cmap="gray")
plt.figure()
plt.imshow(stripes_naive, cmap="gray")
plt.figure()
plt.imshow(stripes_trans, cmap="gray")
```
