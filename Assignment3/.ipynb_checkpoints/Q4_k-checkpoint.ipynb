```python
from numpy import fft
from skimage import io, util, color, measure
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
```

```python
img_dir = "./images/"
```

## 1.

```python
lena = io.imread(img_dir + 'lena.png')[:, :, 0]
cm = io.imread(img_dir + 'cameraman.png')

F = fft.fft2(lena)
H = fft.fft2(cm)

inv = np.abs(fft.ifft2(np.multiply(F, H)))
print(inv.shape)
conv = convolve2d(lena, cm.astype(np.float))[127:383, 127:383]
print(conv.shape)

plt.figure()
plt.imshow(np.abs(inv), cmap="gray")

plt.figure()
plt.imshow(conv, cmap="gray")

mse = np.mean((conv - inv) ** 2)
print(mse)
```

## 2.

```python
lena = io.imread(img_dir + 'lena.png')[:, :, 0]
bear = io.imread(img_dir + 'bear.jpg')[:, :, 0]
M = lena.shape[0] + bear.shape[0] - 1
N = lena.shape[1] + bear.shape[1] - 1

fil = (1/25.0) * np.ones((5, 5))
lena_pad = np.zeros((M, N), dtype=np.uint8)
lena_pad[:lena.shape[0], :lena.shape[1]] = lena
bear_pad = np.zeros((M, N), dtype=np.uint8)
bear_pad[:bear.shape[0], :bear.shape[1]] = bear
conv = convolve2d(lena, bear.astype(np.float))
print(conv.shape)
plt.figure()
plt.imshow(conv, cmap="gray")
```

```python
M = lena.shape[0] + bear.shape[0] - 1
N = lena.shape[1] + bear.shape[1] - 1

res = fft.ifft2(fft.ifftshift(fft.fftshift(fft.fft2(lena, s=(M, N))) * fft.fftshift(fft.fft2(bear, s=(M, N)))))
res = np.abs(res)
print(res.shape)
plt.figure()
plt.imshow(res, cmap="gray")
print(np.mean((res - conv) ** 2))
```
